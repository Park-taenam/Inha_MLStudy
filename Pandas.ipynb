# DataFrame
- 2차원 이하의 데이터들만 DataFrame으로 변환될 수 있음.

## 1차원

```python
# 1차원 list -> ndarray -> DataFrame
lst1 = [1,2,3]

arr1 = np.array(lst1)
print(arr1)
print(type(arr1), '\n')

df1 = pd.DataFrame(lst1, columns = ['col1'])
print(df1)
print(type(df1), '\n')


df2 = pd.DataFrame(arr1, columns = ['col1'])
print(df2)
print(type(df2))
```

## 2차원

```python
# 2차원 데이터 -> DataFrame
lst2 = [[1,2,3],
       [11,12,13]]

arr2 = np.array(lst2)
print(arr2)
print(type(arr2), '\n')

df3 = pd.DataFrame(lst2, columns = ['col1', 'col2', 'col3'])
print(df3)
print(type(df3), '\n')

df4 = pd.DataFrame(arr2, columns = ['col1','col2','col3'])
print(df4)
print(type(df4))
```

## Dictionary

```python
dict1 = {'col1' : [1,11], 'col2' : [2,22], 'col3' : [3,33]}
df_dict = pd.DataFrame(dict1)
print(df_dict)
print(type(df_dict))
```

## DataFrame -> ndarray

```python
arr3 = df_dict.values
print(arr3)
print(type(arr3))
```

## DataFrame -> list, dict

```python
lst3 = df_dict.values.tolist()
print(lst3)
print(type(lst3), '\n')

dict3 = df_dict.to_dict('list')
print(dict3)
print(type(dict3))
```

## iloc[], ioc[]
> iloc[]
- 위치 기반 인덱싱만 허용
- 행과 열 값으로 integer만

> loc[]
- 명칭 기반으로 데이터 추출
- 명칭 기반이기때문에 ':'을 적용할 때 '시작값 ~ 종료값-1"이 아니라 '시작값 ~ 종료값' 이다
- 숫자 형이 아닐 수 있기때문에 -1을 할 수 없음

```python
data = {'Name' : ['Park', 'Kim', 'Lee', 'Baek'],
          'Year' : [2011, 2016, 2015, 2018],
          'Gender' : ['Male', 'Female', 'Male', 'Female']}
data_df = pd.DataFrame(data, index = ['one', 'two', 'three', 'four'])
data_df
```

```python
data_df.reset_index(inplace=True)
data_df.index = data_df.index + 1
data_df.rename(columns = {'index' : 'old_index'})
```

### iloc[]

```python
# iloc
data_df.iloc[0,0]
```

```python
# iloc는 명칭으로 불러올 수 없음
# 오류발생
data_df.iloc[0, 'Name']
```

```python
# index 0,1 만 나옴
data_df.iloc[0:2, 0]
```

### loc[]

```python
# loc
# index명이 0은 없기때문에 오류나옴
data_df.loc[1, 'Name']
```

```python
# 1,2 다 나옴
data_df.loc[1:2, 'Name']
```

# Apply lambda
- 세분화를 많이하려면 함수를 따로 만들어서 사용하는게 더 효율적일수 있다

```python
def get_square(a):
    return a**2

print(get_square(3))
```

```python
lambda_square = lambda x : x**2
print(lambda_square(3))
```

```python
a = [1,2,3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head()
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head()
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df[['Age', 'Age_cat']].head()
```
